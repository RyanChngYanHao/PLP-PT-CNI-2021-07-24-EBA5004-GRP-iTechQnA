## 4. Download non python-tagged questions from Stack API (10% as sample)

```python
import json

from datetime import datetime, timedelta
from stackapi import StackAPI

SITE = StackAPI('stackoverflow', key='2eveOGZYZCywkWwApnSqIw((')
SITE.max_pages = 100
```

```python
# based on 2018 post
# https://towardsdatascience.com/the-frequency-of-tags-on-stack-overflow-2fb47600e2b2
top20 = ['java*', 'c#', 'php', 'android', 
         'jquery', 'html', 'c++', 'ios', 
         'css', 'mysql', 'sql', '*.net',
         '*rails', 'objective-c', 'c', 'arrays', 'angularjs']
```

```python
def write_json3(start, end, time_length, tag): 
    """
    start: datetime
    end: datetime
    time_length: timedelta    """
    
    this_d = start
    this_d_plus1 = this_d + timedelta(1)    
    
    while this_d < end:
        for i in range(len(tag)):
            # cap at 10,000 entries per fetch, 10,000 calls per day

            print(f'Begin fetching {tag[i]} from {this_d} to {this_d_plus1}...')
            o_meta = SITE.fetch('questions', 
                                fromdate=this_d, todate=this_d_plus1, 
                                tagged=tag[i],
                                nottagged='python;python-*',
                                filter='withbody')        

            print(f'Quota remaining == {o_meta["quota_remaining"]}')
            filename = f'./data/o{i}_{this_d.strftime("%Y%m%d")}_{this_d_plus1.strftime("%Y%m%d")}.txt'    
            with open(filename, 'w') as convert_file:
                 convert_file.write(json.dumps(o_meta))
                    
            print(f'\t{filename} successfully saved')
            
        this_d += time_length
        this_d_plus1 = this_d + timedelta(1)           
```

```python
%%time
write_json3(datetime(2018,1,1), datetime(2020,1,1), timedelta(10), top20)
```

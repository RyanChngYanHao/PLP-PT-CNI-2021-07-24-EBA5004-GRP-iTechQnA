# 1. Download python-tagged questions from Stack API

```python
import json

from datetime import datetime, date, time, timedelta
from stackapi import StackAPI

SITE = StackAPI('stackoverflow', key='2eveOGZYZCywkWwApnSqIw((')
SITE.max_pages = 100
```

```python
def write_json(start, end, time_length): 
    """
    start: datetime
    end: datetime
    time_length: timedelta
    """
    
    this_d = start
    next_d = start + time_length  
    
    while this_d < end:
        # cap at 10,000 entries per fetch, 10,000 calls per day
        
        print(f'Begin fetching {this_d} to {next_d}...') 
        q_meta = SITE.fetch('questions', 
                            fromdate=this_d, todate=next_d, 
                            tagged='python',
                            filter='withbody')        
        
        print(f'Quota remaining == {q_meta["quota_remaining"]}')
        filename = f'./data/q{this_d.strftime("%Y%m%d")}_{next_d.strftime("%Y%m%d")}.txt'    
        with open(filename, 'w') as convert_file:
             convert_file.write(json.dumps(q_meta))
        this_d += time_length
        next_d += time_length         
        
        print(f'\t{filename} successfully saved')  
```

```python
%%time
write_json(datetime(2018,1,1), datetime(2020,1,1), timedelta(10))
```

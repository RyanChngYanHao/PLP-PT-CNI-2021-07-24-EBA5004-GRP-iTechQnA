# 3. Download answers from questions from Stack API

```python
import json
import pandas as pd

from datetime import datetime, date, time, timedelta
from stackapi import StackAPI

SITE = StackAPI('stackoverflow', key='2eveOGZYZCywkWwApnSqIw((')
SITE.max_pages = 100
```

```python
def write_json2(qids, id_length):
    """
    qids: pandas series
    id_length: int
    """
    s = 0; total = len(qids)
    while s < total:
        e = min(s + id_length, total)        
        # cap at 100 entries per fetch
        
        print(f'Retrieving answers of questions {s} to {e}')
        a_meta = SITE.fetch('questions/{ids}/answers', ids=qids[s:e],
                            tagged='python',
                            filter='withbody')
        print(f'Quota remaining == {a_meta["quota_remaining"]}')
        
        head = min(qids[s:e]); tail = max(qids[s:e])
        filename = f'./data/a{head}_{tail}.txt'    
        with open(filename, 'w') as convert_file:
             convert_file.write(json.dumps(a_meta))                
        s += id_length
        
        print(f'\t{filename} successfully saved') 
```
